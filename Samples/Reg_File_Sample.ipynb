Importing needed libraries 

```python
from astropy import units
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from AstroDC import AstroDC
```

Loading and preparing data Using AstroDC that changes the type of data from XML+CSV to CSV:

```python
AstroDC.Vizier_XMLCSV_to_CSV("/content/AstroDC/Data/vizier_votable_Ukirt_NGC6822.tsv","Ukirt_NGC6822")
```

```python
Ukirt_NGC6822 = pd.read_csv ('/content/Ukirt_NGC6822.csv').replace(r'^\s*$', np.nan, regex=True)
Ukirt_NGC6822
```

Making Reg file using AstroDC.Reg_file : 

```python
AstroDC.Reg_file(Ukirt_NGC6822 ,'_RAJ2000(deg)','_DEJ2000(deg)', Units = units.deg)
```

Open and read the file after the appending:

```python
f = open("reg_file.reg", "r")
print(f.read())
```
